Homework 8: CSV Data Manipulation with Pandas Dataframe

Step 1: load the csv data

```python
import pandas as pd
patient_df = pd.read_csv("outpatient_sample.csv")
patient_df.head()
```

Step 2: New sub-dataframe with "Patient ID" and "Visit ID" information, get Total Visits.

```python
# get both Patient_ID and Visit_ID columns
new_df = patient_df.loc[:, ["Patient_ID", "Visit_ID"]]
# group by patient id
visit_df = new_df.groupby(["Patient_ID"])
# for each patient id, count number of visits
totalvisit_df = visit_df.agg({'Visit_ID' : 'count'})
# rename to required column name
totalvisit_df.columns = ['Total_Visits']
totalvisit_df
```

Step 3: New sub-dataframe with "Patient ID" and all physician related columns information, get Total Physicians.

```python
# function to get more than one physician columns
def get_unique(group):
    # all three physician related columns
    physician_cols = [group['Primary_Physician'], 
                      group['Operating_Physician'], 
                      group['Other_Physician']]
    # return length of unique physicians
    return len(pd.concat(physician_cols).dropna(inplace=False).unique())

# group by patient id with help of get_unique
physician_gp = patient_df.groupby("Patient_ID").apply(get_unique)
# pandas series to dataframe
physician_gp = physician_gp.to_frame()
# rename to required column output name
physician_gp.columns = ['Total_Physicians']
physician_gp
```

Step 4: New sub-dataframe with "Patient ID" and all ICD9 Diagnosis related columns information, get Total Diagnosis.

```python
# function to get more than one diagnosis columns
def get_diagnosis(group):
    diagnosis_cols = [group['ICD9_DGNS_CD_1'], 
                      group['ICD9_DGNS_CD_2'], 
                      group['ICD9_DGNS_CD_3'],
                      group['ICD9_DGNS_CD_4'],
                      group['ICD9_DGNS_CD_5'],
                      group['ICD9_DGNS_CD_6'],
                      group['ICD9_DGNS_CD_7'],
                      group['ICD9_DGNS_CD_8'],
                      group['ICD9_DGNS_CD_9'],
                      group['ICD9_DGNS_CD_10']]
    # return length of unique diagnosis
    return len(pd.concat(diagnosis_cols).dropna(inplace=False).unique())

# group by patient id with help of get_diagnosis
diagnosis_gp = patient_df.groupby("Patient_ID").apply(get_diagnosis)
# pandas series to dataframe
diagnosis_gp = diagnosis_gp.to_frame()
# rename to required column output name
diagnosis_gp.columns = ['Total_Diagnosis']
diagnosis_gp
```

```python
# function to get more than one diagnosis columns
def get_most_freq(group):
    diagnosis_cols = [group['ICD9_DGNS_CD_1'], 
                      group['ICD9_DGNS_CD_2'], 
                      group['ICD9_DGNS_CD_3'],
                      group['ICD9_DGNS_CD_4'],
                      group['ICD9_DGNS_CD_5'],
                      group['ICD9_DGNS_CD_6'],
                      group['ICD9_DGNS_CD_7'],
                      group['ICD9_DGNS_CD_8'],
                      group['ICD9_DGNS_CD_9'],
                      group['ICD9_DGNS_CD_10']]
    # return value name of unique diagnosis
    return pd.concat(diagnosis_cols).dropna(inplace=False).value_counts().idxmax()

# group by patient id with help of get_most_freq
diagnosis_freq = patient_df.groupby("Patient_ID").apply(get_most_freq)
# pandas series to dataframe
diagnosis_freq = diagnosis_freq.to_frame()
# rename to required column output name
diagnosis_freq.columns = ['Most_Freq_Diganosis']
diagnosis_freq
```

```python
# merge all together into a final master dataframe
merge_one = pd.merge(totalvisit_df, physician_gp, on="Patient_ID")
merge_two = pd.merge(merge_one, diagnosis_gp, on="Patient_ID")
final_merge = pd.merge(merge_two, diagnosis_freq, on="Patient_ID")

final_merge
```
